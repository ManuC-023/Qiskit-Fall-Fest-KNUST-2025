```python
# üí∞ TASK: Quantum Random Number Generator (QRNG)
# Category: Quantum Cryptography
# Concept: Use a qubit in superposition to generate true randomness.
# Why it‚Äôs real: Random numbers are essential for security, passwords, and encryption keys.
# Tools: Qiskit + Hadamard gate
# ‚≠ê (Easy)
# Challenge Idea:
# Generate 100 random quantum bits, plot how evenly distributed they are, and compare with Python‚Äôs random module.
# non entaglement circuit using the qiskit lib
# write a short note about the project you did

# qc  = ..........?   #Create your quantum circuit with some qubits
# ...........   # apply a hadamard gate on your qubit system

NBITS = 100
qc = QuantumCircuit(1)              # Quantum circuit with 1 qubit and 1 classical bit
qc.h(0)                             # Apply Hadamard gate to put qubit in superposition 
qc.measure_all()                    # Measure to collapse the superposition to classical bits
qc.draw(output='mpl')               # Draw the circuit
```

```python
# TASK 2: Now let's perform a simulation on our custom quantum circuit

# sampler = .............?
# result = ....?.run([qc] ,shots=1024).result()
# print(.....?[0].data.meas.get_counts())
# .........?(result[0].data.meas.get_counts())  #plotting graph of the output

sampler = StatevectorSampler()
result = sampler.run([qc], shots=NBITS).result()
print("Quantum random bits counts:", result[0].data.meas.get_counts())
plot_histogram(result[0].data.meas.get_counts())    #plotting graph of the output

# plot_histogram(results)
```

```python
# Compare with Python's random module

classic_bits = [np.random.randint(0, 2) for n in range(NBITS)]                  # Random classical bits using numpy

classic_bits_counts = {'0': classic_bits.count(0), '1': classic_bits.count(1)}  # Creates dicitionary with counts of 0s and 1s
print("Classical random bits counts:", classic_bits_counts)
plot_histogram(classic_bits_counts)                                             # Plotting graph of the output (classical bits)
```
